```python
from diffusers import StableDiffusionPipeline

import torch
```

```python
model_name = 'dreamlike-art/dreamlike-photoreal-2.0'

pipe = StableDiffusionPipeline.from_pretrained(model_name, torch_dtype =torch.float16)
if torch.cuda.is_available():
    print("CUDA is available, moving model to GPU.")
    pipe = pipe.to('cuda')
else:
    print("CUDA is not available, moving model to CPU. Performance may be significantly slower.")
    pipe = pipe.to('cpu')
```

```python
prompts = ['Data analysis using power bi cover',
           'futuristic city skyline at night, neon lights, cyberpunk vibe',
           "calm beach scene with gentle waves, sunrise, warm tones, minimalistic  "]
```
